```python
import pandas as pd
import numpy as np
import math
```

```python
X_train = pd.read_csv('C:\\Users\himab\Documents\spring2020\ML\hw2\hw2-data\Gaussian_process\X_train.csv', header = None)
y_train = pd.read_csv('C:\\Users\himab\Documents\spring2020\ML\hw2\hw2-data\Gaussian_process\y_train.csv', header = None)
X_test = pd.read_csv('C:\\Users\himab\Documents\spring2020\ML\hw2\hw2-data\Gaussian_process\X_test.csv', header = None)
y_test = pd.read_csv('C:\\Users\himab\Documents\spring2020\ML\hw2\hw2-data\Gaussian_process\y_test.csv', header = None)
```

```python
#Q1
def Rij(list_1,list_2,b):
    a = []
    s = 0
    for i in range(len(list_1)):
        s += pow((list_1[i]-list_2[i]),2)
    return math.exp(-s/b)

print(Rij([1,2,3],[1,2,3],2))
            
        
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
def Kn(X_train,b):
    list1 = []
    for i in range(len(X_train)):
        list2 = []
        for j in range(len(X_train)):
            list2.append(Rij(X_train[i],X_train[j],b))
        list1.append(list2)
    return list1
```

```python
def KxDn(x,Dn,b):
    list1 = []
    for j in range(len(Dn)):
        list1.append(Rij(x,Dn[j],b))
    return list1
```

```python
def mu(x,X_train,b,sig_sq, y):
    list1 = []
    for i in range(len(x)):
        a = np.array(KxDn(x[i],X_train,b))
        second_term = (sig_sq * np.identity(len(X_train))) + Kn(X_train,b)
        second_term_inv = np.linalg.inv(second_term)
        mat_mul1 = np.matmul(a,second_term_inv)
        list1.append(np.matmul(mat_mul1,y))
    return(list1)
```

```python
print(mu(X_test,X_train,3,1.4,y_train))
```

```python
#Part (b) #each entry of x_test
b_values = [5,7,9,11,13,15]
sig_sq_values = [.1,.2,.3,.4,.5,.6,.7,.8,.9,1]
means = {}
for b in b_values:
    means[b] = {}
    for sig_sq in sig_sq_values:
        means[b][sig_sq] = mu(X_test,X_train,b,sig_sq,y_train)
means
        
```

```python
def getRMSEvalues(y_test,mu_value):
    val = y_test - mu_value
    mult = val **2
    rmse = math.sqrt(sum(mult)/42)
    return(rmse)
l = {}
rmse_values = {}
for b in b_values:
    l[b] = {}
    rmse_values[b] = {}
    for var in sig_sq_values:
        l[b][var] = mu(X_test,X_train, b, var,y_train)
        rmse = getRMSEvalues(y_test,l[b][var])
        rmse_values[b][var] = rmse
```

```python
print('RMSE for given variance and b values')
table = pd.DataFrame(rmse_values)
table
```
